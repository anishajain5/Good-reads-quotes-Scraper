```python
def get_quotes(url):
    res= requests.get(url)
    soup= BeautifulSoup(res.text)
    quote_divs= soup.find_all("div", attrs={"class":"quote"})
    quotes=[]

    for quote_div in quote_divs:
        quote_divs1= quote_div.find_next("div", attrs={"class":"quoteText"})
        striped=quote_divs1.text.strip()
    
        striped_1= striped.split("\n")
        quote= striped_1[0][1:-1]
        author= striped_1[-1].strip()
        left_div=quote_div.find_next("div", attrs={"class":"left"})
        tag= [tag.text for tag in left_div.find_all("a")]
    
        quote_item= {"text": quote, "author": author, "tags": tag}
    
        
        quotes.append(quote_item)
    return quotes
    
```
